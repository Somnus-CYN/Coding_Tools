<font color=black size=5 face=雅黑>**VScode连接github**</font>
git是一种常见的管理代码的工具，下面介绍如何将本地的代码上传至Github。
1. 首先在Github上新建一个仓库，或者导航到您要上传文件夹的现有仓库。

2. 打开VSCode，确保您已经安装并配置了Git工具。
3. 在VSCode的侧边栏中，打开您想要上传到GitHub的文件夹。
4. 在VSCode的顶部菜单中选择“终端(Terminal)” -> “新建终端(New Terminal)”来打开集成终端。
5. 在终端中，输入以下命令以初始化Git仓库：

git init

6. 添加文件到Git仓库，可以使用以下命令将所有文件添加到暂存区：

git add .

或者，如果只想添加特定的文件，可以使用以下命令：

git add 文件名

7. 提交更改，使用以下命令：

git commit -m "提交说明"

8. 在GitHub上创建的仓库中获取远程仓库的URL。您可以通过点击“Code”按钮并选择HTTPS或SSH获取URL。
9. 将本地仓库与远程仓库关联，使用以下命令：

git remote add origin 远程仓库的URL

10. 将本地提交推送到GitHub，使用以下命令：

git push -u origin master

如果您在GitHub上使用的是不同的分支名称，请将"master"替换为您的分支名称。
